## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty

df['subtotal'] = df['unit_price'] * df['qty']
df.head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

df['order_weight'] = df['unit_weight'] * df['qty']

def shipping_rate_calc(weight):
    if weight > 50:
        return 7.00
    elif weight <= 50:
        return 10.00

def shipping_price_calc(weight):
    if weight > 50:
        return weight * 7
    elif weight <= 50:
        return weight * 10

df['shipping_rate'] = df['order_weight'].apply(shipping_rate_calc)
df['shipping_price'] = df['order_weight'] * df['shipping_rate']

# df['shipping_price'] = df['order_weight'].apply(shipping_price_calc)

df.head()
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
df['sales_tax'] = (df['subtotal'] + df['shipping_price']) * 0.0925
df['total_price'] = df['subtotal'] + df['shipping_price'] + df['sales_tax']
df.head()


```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

df['total_unit_cost'] = df['unit_cost'] * df['qty']
df['total_cost'] = df['total_unit_cost'] + df['shipping_price']
df.head()
```

```python
# Create a column for the profit of
# each line using line cost and line price

df['profit'] = df['total_price'] - df['total_cost']
df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.

def client_id_total_price_calc(client_id):
    return df.loc[df['client_id']== client_id, 'total_price'].sum()

for client_id in top_5_clients:
    total_spent = client_id_total_price_calc(client_id)
    print(f'Client ID {client_id} spent ${total_spent:,.2f}.')
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.

def client_id_total_qty_calc(client_id):
    return df.loc[df['client_id'] == client_id, 'qty'].sum()

def client_id_total_shipping_price_calc(client_id):
    total_weight = df.loc[df["client_id"] == client_id,"unit_weight"]* df.loc[df['client_id'] == client_id, 'qty']
    shipping_price =total_weight.apply(lambda weight:weight * 7 if weight > 50 else weight * 10)
    return shipping_price.sum()

def client_id_total_revenue_calc(client_id):
    return df.loc[df["client_id"] == client_id,"total_price"].sum()
#total profit by client - identify line cost, then subtract cost from total price
def client_id_total_profit_calc(client_id):
    line_cost = df.loc[df["client_id"] == client_id, "unit_cost"] * df.loc[df['client_id'] == client_id, 'qty'] + df.loc[df["client_id"]== client_id, "shipping_price"]
    line_profit = df.loc[df["client_id"] == client_id, "total_price"] - line_cost
    return line_profit.sum()

summary_list = []

for client_id in top_5_clients:
    qty_summary = client_id_total_qty_calc(client_id)
    shipping_price_summary = client_id_total_shipping_price_calc(client_id)
    revenue_summary = client_id_total_revenue_calc(client_id)
    profit_summary = client_id_total_profit_calc(client_id)
    summary_data = {
        "client_id":client_id,
        "qty": qty_summary,
        "shipping_price": shipping_price_summary,
        "revenue": revenue_summary,
        "profit": profit_summary
    }
    summary_list.append(summary_data)
    
summary_df = pd.DataFrame(summary_list)
summary_df = summary_df.sort_values(by = "profit", ascending = False)
summary_df

```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.

summary_df['shipping_price'] = summary_df['shipping_price']/1000000
summary_df['revenue'] = summary_df['revenue']/1000000
summary_df['profit'] = summary_df['profit']/1000000

summary_df = summary_df.rename(columns={

    'client_id': 'Client ID',
    'qty':'Quantity',
    'profit':'Total Profit (Millions)',
    'revenue':'Total Revenue (Millions)',
    'shipping_price':'Total Shipping Price (Millions)'
}
)

# Sort the updated data by "Total Profit" form highest to lowest
summary_df=summary_df.sort_values(by = "Total Profit (Millions)", ascending=False)

summary_df['Total Shipping Price (Millions)'] = summary_df['Total Shipping Price (Millions)'].map('${:,.2f}'.format)
summary_df['Total Revenue (Millions)'] = summary_df['Total Revenue (Millions)'].map('${:,.2f}'.format)
summary_df['Total Profit (Millions)'] = summary_df['Total Profit (Millions)'].map('${:,.2f}'.format)

print(summary_df)
```
